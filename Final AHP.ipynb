```python
import pandas as pd

def generate_table(data):
    table = []

    for key, value in data.items():
        if 'elements' in value:
            elements = value['elements']
            for sub_key, sub_value in elements['global_weights'].items():
                sub_criteria_name = f"{key}_{sub_key}"
                global_weight = value['global_weight']
                local_weight = value['local_weight']
                sub_criteria_weight = sub_value
                sub_weights = value['target_weights']

                sub_weights_ordered = [sub_weights.get(str(i), None) for i in range(1, 5)]

                table.append({
                    'Criteria': key,
                    'Global Weight': global_weight,
                    'Local Weight': local_weight,
                    'Sub-Criteria': sub_criteria_name,
                    'Sub-Criteria Weight': sub_criteria_weight,
                    'Sub-Criteria Weight 1': sub_weights_ordered[0],
                    'Sub-Criteria Weight 2': sub_weights_ordered[1],
                    'Sub-Criteria Weight 3': sub_weights_ordered[2],
                    'Sub-Criteria Weight 4': sub_weights_ordered[3]
                })

    return table



# Generate the table
table = generate_table(data)

# Convert the table to a pandas DataFrame
df = pd.DataFrame(table)

# Save the DataFrame to an Excel file
excel_file_path = r"C:\Users\srika\Dropbox\PC\Downloads\Ahp Atm Weights\P10.xlsx"
df.to_excel(excel_file_path, index=False)

print(f"Table saved to '{excel_file_path}'")
```
